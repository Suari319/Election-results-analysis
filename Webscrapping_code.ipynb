```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://results.eci.gov.in/AcResultGen2ndJune2024/statewiseS211.htm"
req=requests.get(url)
```

```python
soup=BeautifulSoup(req.text, "html.parser")
print(soup.prettify())
```

```python
import pandas as pd
```

```python
table=soup.find("table")
print(table)
```

```python
titles=table.find_all("th")
heading=[title.text.strip() for title in titles[1:] if title.text.strip()]
print(heading)
```

```python
df=pd.DataFrame(columns=heading)
```

```python
df
```

```python
column_data = soup.find_all('tr')
```

```python
for j in range(1,3):
    url = f'https://results.eci.gov.in/AcResultGen2ndJune2024/statewiseS21{j}.htm'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    column_data = soup.find_all('tr')
    for i in column_data[2:]:
        row_data = i.find_all('td')
        individual_row_data = []
        for data in row_data:
        # Remove span tags inside td
            for tooltip in data.find_all('div', class_='tooltip'):
                tooltip.decompose()
        # Extract text after removing spans
            for tooltip_icon in data.find_all('span', class_='tooltip-icon'):
                tooltip_icon.decompose()

        # Append non-empty data to the list
            cleaned_data = data.text.strip()
            if cleaned_data:
                individual_row_data.append(cleaned_data)

    # Remove duplicate entries in the same row
        unique_row_data = []
        for item in individual_row_data:
            if item not in unique_row_data:
                unique_row_data.append(item)
    # Ensure the list has the correct number of elements before printing
        if len(unique_row_data) > 1:
            data = unique_row_data
            print(data)
            df.loc[len(df)] = data
```
